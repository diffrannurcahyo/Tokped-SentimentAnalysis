```python
from google_play_scraper import app
import pandas as pd
import numpy as np
```

```python
 #scrape desire number of reviews
 from google_play_scraper import Sort, reviews

 result, continuation_token = reviews(
    'com.tokopedia.tkpd',
    lang= 'id',
    country = 'id',
    sort = Sort.NEWEST,
    count = 10000,
    filter_score_with = None
 )

 #scraping tokopedia
```

```python
#membuat data frame dengan nama 'datacoba'
df = pd.DataFrame(np.array(result),columns=['review'])
df = df.join(pd.DataFrame(df.pop('review').tolist()))

#menampilkan data teratas
df
```

```python
#memastikan jumlah data
len(df.index)
```

```python
#untuk menampilkan nama, skor, tanggal, dan review
df[['userName','score','at','content']].head()
```

```python
#mengurutkan data berdasarkan tanggal
df_scrape = df[['userName','score','at','content']] 
df_sorted = df_scrape.sort_values(by='at', ascending=False) 
df_sorted.head()
```
